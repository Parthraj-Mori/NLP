```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as pt
```

```python
f1 = pd.read_csv("Restaurant_Reviews.tsv", delimiter= "\t", quoting = 3)
```

```python
f1.head()
```

```python
import re
import nltk
```

```python
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
corpus = []
ps= PorterStemmer()
for i in range(0, len(f1)):
    review = re.sub("[^a-zA-Z]", " ", f1["Review"][i])
    print(review)
    review = review.lower()
    review = review.split()
    all_stopword= stopwords.words("english")
    all_stopword.remove('not')
    all_stopword.remove("aren't")
    all_stopword.remove("hasn't")
    all_stopword.remove("haven't")
    review1= []
    for j in review :
        if not j in all_stopword:
            review1.append(ps.stem(j))
            
    review= " ".join(review1)
    corpus.append(review)
            
   
    
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
cv = CountVectorizer(max_features= 1500)
X = cv.fit_transform(corpus).toarray()
Y= f1.iloc[:,-1]
```

```python
len(X[0])
```

```python
X
```

```python
Y=Y
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size= 0.1, random_state= 0)
```

```python
x_test
```

```python
len(x_test)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gn= GaussianNB()
```

```python
gn.fit(x_train,y_train)
```

```python
y_pred= gn.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test , y_pred)
```
